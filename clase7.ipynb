```python
path = "/content/drive/MyDrive/Colab Notebooks/Canada.xlsx"
```

```python
import pandas as pd
```

```python
df = pd.read_excel(path,
                   sheet_name = "Canada by Citizenship",
                   skiprows =range(20),
                   skipfooter = 2,
                   engine ="openpyxl"
                  )
```

```python
to_drop = ["Type","Coverage","AREA","REG","DEV","DevName"]
df = df.drop(to_drop, axis =1)
df.head()
```

```python
df.head(2)
```

```python
df["Total"] = df.iloc[:,3:].sum(axis=1)
df.head(2)
```

```python
all(isinstance(column, str)for column in df.columns)
```

```python
df.columns=list(map(str, df.columns))
```

```python
all(isinstance(column, str)for column in df.columns)
```

```python
df.columns
```

```python
x = range (1980,2014)
years = list(map(str,x))
years
```

##Como fue el comportamiento de immigracion desde Italia a Canada durante 1980-2013

```python
df.rename(columns = {"OdName":"Country",
                     "AreaName": "Continent",
                     "RegName": "Region"},
          inplace = True)
df.head(2)
```

```python
df.loc[0,years]
```

```python
df.set_index("Country", inplace=True)
df.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
df.loc["Italy", years]
```

```python
df.loc["Italy", years].plot(kind = "line")
plt.xlabel("Years", fontsize = 14)
plt.ylabel("# of imigrants", fontsize = 14)
plt.title("Immigration pattern from Italy to Canada", fontsize = 17)
plt.text(0,2100,"Cambio de gobierno")
```

```python
dfci=df.loc[["India", "China"],years].transpose()
```

```python
dfci.plot(kind="line")
plt.xlabel("Years", fontsize = 14)
plt.ylabel("# of imigrants", fontsize = 14)
plt.title("Immigration pattern from Italy to Canada", fontsize = 17)
```
